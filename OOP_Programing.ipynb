```python
class Student:
    def __init__(self, name, age, year, homework, mid_term, final):
        self.name = name
        self.age = age
        self.year = year
        self.homework = homework
        self.mid_term = mid_term
        self.final = final

    def get_sum(self):
        return self.homework + self.mid_term + self.final

    def get_grade(self):
        total = self.get_sum()
        if total < 50:
            return "F"
        elif total < 60:
            return "D"
        elif total < 70:
            return "C"
        elif total < 80:
            return "B"
        else:
            return "A"

    def __str__(self):
        return (
            f"{self.name} (Year {self.year})\n"
            f"Homework: {self.homework}\n"
            f"Midterm: {self.mid_term}\n"
            f"Final: {self.final}\n"
            f"Total: {self.get_sum()} → Grade: {self.get_grade()}"
        )
```

```python
name = input("Student's name : ")
age = int(input("Student's age : "))
year = int(input("Student's year : "))

homework = int(input("Homework score : "))
mid_term = int(input("Mid term : "))
final = int(input("Final : "))

stc1 = Student(name, age, year, homework, mid_term, final)


print(stc1)
```
