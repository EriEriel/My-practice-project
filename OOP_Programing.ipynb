```python
class student:
    def __init__(self, name, age, year):
        self.name = name
        self.age = age
        self.year = year
    def __str__(self):
        return f"{self.name} ({self.age}, year {self.year})"


class student_score(student):
    def __init__(self, name, age, year, homework, mid_term, final):
        super().__init__(name, age, year)
        self.homework = homework
        self.mid_term = mid_term
        self.final    = final
    def get_sum(self):
        return(self.homework + self.mid_term + self.final)
    def get_grade(self):
        sum_score = self.get_sum()
        if sum_score < 50:
            self.grade = "F"
        elif 50 <= sum_score < 60:
            self.grade = "D"
        elif 60 <= sum_score < 70:
            self.grade = "C"
        elif 70 <= sum_score < 80:
            self.grade = "B"
        else:
            self.grade = "A"
        return self.grade 
    def __str__(self):
        return(
               f"{str(self.name)} Total score is {str(self.get_sum())} \n"
               f"Homework  : {str(self.homework)} \n"
               f"Mid term  : {str(self.mid_term)} \n" 
               f"Final     : {str(self.final)} \n"
               f"Get grade : {self.get_grade()}")
```

```python
name = input("Student's name : ")
age = int(input("Student's age : "))
year = int(input("Student's year : "))

homework = int(input("Homework score : "))
mid_term = int(input("Mid term : "))
final = int(input("Final : "))

stc1 = student_score(name, age, year, homework, mid_term, final)

print()
print(stc1)
```
